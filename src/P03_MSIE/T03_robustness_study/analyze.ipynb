```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns
```

```python
run_names = [
    "20251117_105629",
    "20251117_105654",
    "20251117_105739",
    "20251117_105941",
    "20251117_110017",
]
steps = [37009, 103088, 214435, 115397, 99513]

df_mon_array = []
df_train_array = []
df_exp = []
for run, step in zip(run_names, steps):
    df_mon_temp = (
        pd.read_csv(f"logs/{run}_monitor.csv", skiprows=1)
        .reset_index()
        .rename(columns={"index": "idx"})
    )
    df_mon_temp["idx"] = df_mon_temp["idx"] + 1  # start from 1
    df_mon_temp.set_index("idx", inplace=True)
    df_mon_temp["accumulated_steps"] = df_mon_temp["l"].cumsum()
    df_mon_temp["run"] = run
    df_mon_array.append(df_mon_temp)

    df_train_temp = pd.read_csv(f"logs/{run}_training.csv")
    df_train_temp["run"] = run
    df_train_array.append(df_train_temp)
    
    df_exp_temp = pd.read_excel(f"saved_models/R_{run}/sol_{step}_exp.xlsx")
    df_exp_temp["run"] = run
    df_exp.append(df_exp_temp)

df_mon = pd.concat(df_mon_array)
df_train = pd.concat(df_train_array)
df_exp = pd.concat(df_exp)
```

```python
data_array = []

for idx, run in enumerate(run_names):
    
    
    print("====================================")
    filt_run = df_exp["run"] == run_names[idx]
    reward = df_exp[filt_run]["reward"].sum()
    best_solution = df_exp[filt_run]["best_solution"].min()

    print(reward)
    filt_run = df_mon["run"] == run_names[idx]
    filt_reward = (df_mon["r"] - reward).abs()< 0.0001
    display(df_mon[filt_run & filt_reward])
    episode = df_mon[filt_run & filt_reward].index.values[0]
    time_elapsed = df_mon[filt_run & filt_reward]["t"].values[0]

    step = steps[idx]
    print(step)
    filt_run = df_train["run"] == run_names[idx]
    filt_step = (df_train["time/total_timesteps"] - step).abs() < 1000
    display(df_train[filt_run & filt_step])
    
    data = {
        "run": run_names[idx],
        "best_solution": best_solution,
        "step": step,
        "reward": reward,
        "episode": episode,
        "time_elapsed": time_elapsed,
    }
    data_array.append(data)
df_summary = pd.DataFrame(data_array)
```

```python
df_summary["time_elapsed_min"] = (df_summary["time_elapsed"] / 60.0).round(2)
df_summary["best_solution"] = df_summary["best_solution"].round(2)
df_summary
df_summary.to_excel("tables/robustness_study_summary.xlsx", index=False)
```
