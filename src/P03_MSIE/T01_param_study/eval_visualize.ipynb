```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import os
import numpy as np

CURRENT_DIR = os.getcwd()
```

```python
filenames = os.listdir(f"{CURRENT_DIR}/eval_results")
df_array = []
for filename in filenames:
    filepath = f"{CURRENT_DIR}/eval_results/{filename}"
    with open(filepath, "rb") as f:
        data_stored = pickle.load(f)
    ds = pd.DataFrame.from_dict(data_stored)
    ds["filename"] = filename
    df_array.append(ds)

ds = pd.concat(df_array)
```

```python
ds
```

```python
fig, ax = plt.subplots(figsize=(14, 5)) 
sns.boxplot(data=ds, x="model_name", y="best_solution", hue="model_name", ax=ax)
plt.title("Comparison of Best Solutions by Agent Type")
plt.xlabel("Agent Type")
plt.ylabel("Best Solution Value")
# Draw horizontal lines for best solutions
colors = sns.color_palette("Set2", 2)
plt.axhline(y=191.81, color=colors[0], linestyle='--', alpha=0.7)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right") 
```

```python
ds["model_name_2"] = ds["model_name"].apply(lambda x: "Fixed-Param Agent" if x != "RL" else "RL Agent")
fig, ax = plt.subplots(figsize=(6, 4)) 
sns.boxenplot(data=ds, x="model_name_2", y="best_solution", hue="model_name_2", ax=ax)
# Draw horizontal lines for best solutions
colors = sns.color_palette("Set2", 2)
plt.axhline(y=191.81, color=colors[0], linestyle='--', alpha=0.7)

# plt.title("Comparison of Best Solutions by Agent Type")
plt.xlabel("Agent Type")
plt.ylabel("Best Solution Value (km)")
# ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right") 

# Annotate best solution line
plt.text(0.45, 205, "Best Solution from RL Agent (191.81 km)", color='black', fontsize=9, va='bottom', ha='left', backgroundcolor='white')


plt.tight_layout()
plt.savefig("figures/best_solution_comparison_boxen.png", dpi=300, bbox_inches='tight')
```

```python
models = ds["model_name"].unique()
models = models[:10]

fig, ax = plt.subplots(figsize=(6, 4))

# colors = sns.color_palette("husl", len(models))
colors = sns.color_palette("Set2", len(models))
for model, color in zip(models, colors):
    filt = ds["model_name"] == model
    for history in ds[filt]["global_solution_history"]:
        sns.lineplot(data=np.array(history), color=color, alpha=0.3, ax=ax)

# Draw horizontal lines for best solutions
plt.axhline(y=191.81, color=color, linestyle='--', alpha=0.7)

# Annotate best solution line
plt.text(5, 205, "Best Solution from RL Agent (191.81 km)", color='black', fontsize=9, va='bottom', ha='left', backgroundcolor='white')

# Annotate the end of each model's lines
plt.text(2000, 550, "Fixed-Param Agent", color="black", fontsize=11, va='center', ha='left', backgroundcolor='white')
plt.text(2000, 300, "RL Agent", color="black", fontsize=11, va='center', ha='left', backgroundcolor='white')



# plt.title("Solution Improvement Over Time by Agent Type")
plt.xlabel("Iteration")
plt.ylabel("Global Solution Value (km)")
plt.tight_layout()
plt.savefig("figures/solution_improvement_over_time.png", dpi=300, bbox_inches='tight')
```
