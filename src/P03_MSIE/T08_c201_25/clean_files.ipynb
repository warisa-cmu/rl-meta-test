```python
import pathlib
import os
import re
```

```python
# current_dir = (pathlib.Path(__file__).parent.resolve(),)
current_dir = os.getcwd()
saved_models_dir = os.path.join(current_dir, "saved_models")
print(saved_models_dir)
for folder in os.listdir(saved_models_dir):
    folder_path = os.path.join(saved_models_dir, folder)
    print(f"Folder: {folder_path}")

    for file in os.listdir(folder_path):
        iteration = re.findall(r"\d+", file)
        if iteration and int(iteration[0]):
            print(iteration)
```
