# Verify state saving and loading for reproducibility

```python
import numpy as np
```

```python
rng1 = np.random.default_rng(seed=42)
```

```python
rng_state = rng1.bit_generator.state
```

```python
rng1.random(10)
```

```python
rng2 = np.random.default_rng(seed=42)
rng2.bit_generator.state = rng_state
rng2.random(10)
```
