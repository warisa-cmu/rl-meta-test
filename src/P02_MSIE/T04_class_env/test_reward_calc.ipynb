```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
p = np.linspace(1, 0, 100)
for alpha_patience in [10, 20, 30, 100]:
    patience_factor = np.exp(-alpha_patience * (1.0 - p))
    plt.plot(
        p,
        patience_factor,
        label=f"alpha={alpha_patience}",
    )
plt.xlabel("Progress")
plt.ylabel("Patience Factor")
plt.title("Patience Factor vs Progress for Different Alpha Values")
plt.legend()
plt.grid()
plt.show()
```

```python
epsilon_target = 1e-6  # Prevent division by zero
improvement = (
    1.216281452
    - 1.212157969
)
value = 1.216281452
target_solution = 300/500
close_to_target = 1 / (
    np.abs(value - target_solution) + epsilon_target
)
alpha_target=4
reward_1 = improvement + alpha_target * close_to_target

patience_remaining = 400
alpha_patience = 100
# Scaled by patience
p = float(np.clip(patience_remaining / 400, 0.0, 1.0))
patience_factor = max(0, np.exp(-alpha_patience * (1.0 - p)))
reward_2 = reward_1 * patience_factor
```
