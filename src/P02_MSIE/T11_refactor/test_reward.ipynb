```python
import numpy as np
from matplotlib import pyplot as plt
import itertools
```

```python
target_solution_sc = 0.1
value_sc = np.linspace(1, 0.01, 100)
```

```python
epsilon_target = 1e-6
close_to_target = 1 / (np.abs(value_sc - target_solution_sc) + epsilon_target)
plt.plot(value_sc, close_to_target)
```

```python
for k in range(1, 10, 2):
    d = np.maximum(value_sc - target_solution_sc, 0.0)
    d0 = 0.2
    reward = 1 - np.exp(-k * d0 / (d + d0))
    plt.plot(value_sc, reward, label=f"k={k}")
    plt.legend()
```

```python
for d0 in range(1, 5):
    k = 2
    d = np.maximum(value_sc - target_solution_sc, 0.0)
    reward = 1 - np.exp(-k * d0 / (d + d0))
    plt.plot(value_sc, reward, label=f"d0={d0}")
    plt.legend()
```

```python
itera = itertools.product([2], [0.1, 0.125, 0.25, 0.5, 1])
for p, c in itera:
    d = np.maximum(value_sc - target_solution_sc, 0.0)
    reward = 1 / (1 + np.power((d / c), p))
    plt.plot(value_sc, reward, label=f"p={p}, c={c}")
    plt.legend()
```
